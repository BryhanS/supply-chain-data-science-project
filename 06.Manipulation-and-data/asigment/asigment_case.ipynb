```python
## the monst popular cinty destination city from new youkr
flight.columns
table1 = flight.groupby('dest').agg(count = ('dest','count')).sort_values(by='count',ascending=False).reset_index()
```

```python
### with airline is the most punctual?

flight['total_delay'] = flight['arr_delay'] + flight['dep_delay']
```

```python
flight.columns
```

```python
table2 = flight.groupby('carrier').agg(mean_deplay=('total_delay','mean')).sort_values(by='mean_deplay').reset_index()
```

```python
pd.merge(table2,airlines,how='left')
```

```python
### what destination has the longest duration

flight.air_time.describe()

table3 = flight.groupby(['origin','dest']).agg(avarege_air_time = ('air_time','mean')).reset_index().sort_values(by='avarege_air_time',ascending=False)
```

```python
pd.merge(table3, airports[['faa','name']],how='left',left_on='dest', right_on='faa')
```

```python
### which airline has the highest capacity of seats?
airlines
plane.columns
flight.columns
```

```python
carrier_tailnum = flight[['carrier', 'tailnum']].drop_duplicates()
```

```python
seats = pd.merge(carrier_tailnum, plane[['tailnum','seats']] , how='left').groupby('carrier').agg(total_seats = ('seats','sum')).sort_values(by='total_seats', ascending=False)
seats
```

```python
### which airplane is the highest in use from which manufactres
airplane_use = flight.groupby('tailnum').agg(count = ('tailnum','count')).reset_index()
plane.columns
```

```python
pd.merge(plane[['tailnum','model','manufacturer']],airplane_use).groupby(['model','manufacturer']).agg(total_flight=('count','sum')).sort_values(by='total_flight',ascending=False)
```
