```python

cars = cars.rename(columns={'name': 'car_name'})
cars.columns
```

```python
cars_price = cars[['car_name','Price']]
cars_price.head(6)
```

```python
def pricing_category (price):
    if(price < 20000):
        return 'Budget Car'
    elif (price > 20000 and price < 35000):
        return 'Suitable Car'
    else:
        return 'Expensive car'
```

```python
cars_price['category'] = cars_price['Price'].map(pricing_category)
cars_price.head(5)
```

```python

cars_price['category_2'] = 'NA'

for i in range(cars_price.shape[0]):
    cars_price.category_2[i] = pricing_category(cars_price.Price[i])


cars_price.head(5)
```
